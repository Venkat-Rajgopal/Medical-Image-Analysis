# Analysing Bio_Medical images

```python
import pathlib
import os, random
import imageio                      # reading images
import matplotlib.pyplot as plt
import numpy as np

im_path = pathlib.Path('Data/chest/')

files = []        
for currentFile in im_path.iterdir():  
    im_file = currentFile        
    im = imageio.imread(im_file)
    files.append(im)
    
print('Images in drive', len(files))
```

```python
[i for i in range(len(files))]

# selecting a random image from the directory. 
random.choice(np.arange(len(files)))
```

```python
# Print image attributes
print('Image type:', type(files[0]))
print('Shape of image array:', files[0].shape)
```

```python
# Print the available metadata fields
print(files[0].meta)
```

```python
# print all meta data
im.meta.keys()
```

```python
plt.imshow(files[0], cmap='gray')
plt.axis('off')
plt.show()
```

Images can be drawn with a greater contrast as below. 

```python
# Draw the image with greater contrast
plt.imshow(im, cmap='gray', vmin=-100, vmax=100)
plt.axis('off')
plt.show()
```

```python
print('Sampling', im.meta['sampling'])
print('Modality' , im.meta['Modality'])
print('Sex', im.meta['PatientSex'])
```

## Stacking images, Loading volumes

Stack images of multidimentional data 

```python
# Stacking the whole chest images with np.stack
vol = np.stack([files[0], files[1], files[3]])
print('Dimentional Volume:', vol.shape)
```
